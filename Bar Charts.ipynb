# Bar plot for NMAC

```python
Aircraft = ("2", "4", "8", "16", "32")
plt.style.use('seaborn-v0_8')
Aircraft = (2,4,8,16,32)
NMAC = {
    'Baseline': (0,0,0,1.36,9.96),
    'Action Sheilding': (0,0,0,0.68,2.12),
    'Reward Shaping': (0,0,0,0.24,1.48),
}
x = np.arange(len(Aircraft))  # the label locations
width = 0.3  # the width of the bars
multiplier = 0

plt.rcParams["figure.figsize"] = (4,4)
fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in NMAC.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
#     ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('NMAC', fontsize=13)
ax.set_xlabel('Number of Aircraft', fontsize=13)
ax.set_title(r'\textbf{NMAC Performance Comparison}', fontsize=14)

# plt.savefig('Battery_SOC_low.png', format='png', dpi=140)

ax.set_xticks(x + width, Aircraft)
ax.legend(loc='upper left', ncols=1, fontsize=12)
ax.set_ylim(0, 10.1)

# plt.show()
plt.savefig('NMAC_bar.eps', format='eps', dpi=140)
```
